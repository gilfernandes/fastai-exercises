### Feature Engineering

```python
import functools

class FeatureEngineering():
    
    def __init__(self, train_df, test_df, train_labels_df):
        self.train_df, self.test_df, self.train_labels_df = self.extract_time_features(train_df), self.extract_time_features(test_df), train_labels_df
        self.numerical_columns = ['game_time']
        self.categorical_columns = ['type', 'world']
        self.comp_train_df = self.create_unique_installation_id_df(train_df)
        self.comp_test_df = self.create_unique_installation_id_df(test_df)
        self.agg_operations = ['mean', 'sum', 'min', 'max', 'std', 'skew']
        merge_args = {'left_index':True, 'right_index':True}
        for col in self.numerical_columns:
            self.comp_train_df = self.comp_train_df.merge(self.get_aggregates(self.train_df, col), **merge_args)
            self.comp_test_df = self.comp_test_df.merge(self.get_aggregates(self.test_df, col), **merge_args)
            
        for col in self.categorical_columns:
            self.comp_train_df = self.comp_train_df.merge(self.get_object_columns(self.train_df, col), **merge_args)
            self.comp_test_df = self.comp_test_df.merge(self.get_object_columns(self.test_df, col), **merge_args)
            
        for col in self.categorical_columns:
            self.comp_train_df = self.comp_train_df.merge(
                self.get_numeric_columns_add(self.train_df, col, 'game_time'), **merge_args)
            self.comp_test_df = self.comp_test_df.merge(
                self.get_numeric_columns_add(self.test_df, col, 'game_time'), **merge_args)
            
        self.comp_train_df.reset_index()
        self.comp_test_df.reset_index()
        
        # mode of the title
        labels_map = dict(train_labels_df.groupby('title')['accuracy_group'].agg(lambda x:x.value_counts().index[0]))
        labels = train_labels_df[['installation_id', 'title', 'accuracy_group']]
        labels['title'] = labels['title'].map(labels_map)
        # get title from the test set
        self.comp_test_df['title'] = test_df.groupby('installation_id').last()['title'].map(labels_map).reset_index(drop = True).values
        self.comp_train_df = labels.merge(self.comp_train_df, on = 'installation_id', how = 'left')
        
        
    def create_unique_installation_id_df(self, df):
        comp_df = pd.DataFrame({'installation_id': df['installation_id'].unique()})
        comp_df.set_index('installation_id', inplace = True)
        return comp_df
    
    def extract_time_features(self, df):
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df['date'] = df['timestamp'].dt.date
        df['month'] = df['timestamp'].dt.month
        df['hour'] = df['timestamp'].dt.hour
        df['year'] = df['timestamp'].dt.year
        df['dayofweek'] = df['timestamp'].dt.dayofweek
        df['weekofyear'] = df['timestamp'].dt.weekofyear
        df['dayofyear'] = df['timestamp'].dt.dayofyear
        df['quarter'] = df['timestamp'].dt.quarter
        df['is_month_start'] = df['timestamp'].dt.is_month_start    
        return df
    
    def get_aggregates(self, df: pd.DataFrame, column):
        df_agg = df.groupby('installation_id').agg({column: self.agg_operations})
        df_agg[column].fillna(df_agg[column].mean(), inplace=True)
        df_agg.columns = [f'{column}_{c}' for c in self.agg_operations]
        return df_agg
    
    def get_object_columns(self, df, col):
        df = df.groupby(['installation_id', col])['event_id'].count().reset_index()
        df = df.pivot_table(index = 'installation_id', columns = [col], values = 'event_id')
        df.columns = list(df.columns)
        df.fillna(0, inplace = True)
        return df
    
    def get_numeric_columns_add(self, df, agg_column, column):
        df = df.groupby(['installation_id', agg_column]).agg({f'{column}': self.agg_operations}).reset_index()
        df = df.pivot_table(index = 'installation_id', columns = [agg_column], values = [col for col in df.columns if col not in ['installation_id', agg_column]])
        df[column].fillna(df[column].mean(), inplace = True)
        df.columns = list(df.columns)
        return df
        
    @property
    def comp_train_head(self):
        return self.comp_train_df[:10]
    
    @property
    def comp_test_head(self):
        return self.comp_test_df[:10]

    @property
    def comp_train_shape(self):
        return self.comp_train_df.shape
    
    @property
    def comp_feature_number(self):
        return self.comp_train_x.shape[1]
    
    @property
    def comp_test_shape(self):
        return self.comp_test_df.shape
    
    @property
    def comp_train_x(self):
        return self.comp_train_df[[i for i in self.comp_train_df.columns if i not in ['accuracy_group', 'installation_id']]]
    
    @property
    def comp_train_y(self):
        return self.comp_train_df['accuracy_group']
    
    @property
    def comp_train_ds(self):
        m = self.comp_train_y.size
        x_size = int(m * 0.9)
        return self.comp_train_x[:x_size], self.comp_train_y[:x_size], self.comp_train_x[x_size:], self.comp_train_y[x_size:]
```

## Data Set and Data Loader

```python
from torch.utils.data import DataLoader, Dataset, SequentialSampler, RandomSampler
import torch
from torch import nn

def convert_df_to_tensor(df):
    return torch.tensor(df.values)

class BowlDataSet(Dataset):
    def __init__(self, df_x: pd.DataFrame, df_y: pd.DataFrame):
        self.x = convert_df_to_tensor(df_x).float()
        self.y = convert_df_to_tensor(df_y).long()
        
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, i):
        return self.x[i], self.y[i]
    
```

## Accuracy functions

```python
# quadratic weighted kappa
def qwk3(a1, a2, max_rat=3):
    '''
    a1 - ground truth
    a2 - predicted values
    '''
    assert(len(a1) == len(a2))
    a1 = np.asarray(a1, dtype=int)
    a2 = np.asarray(a2, dtype=int)
    hist1 = np.zeros((max_rat + 1, ))
    hist2 = np.zeros((max_rat + 1, ))
    o = 0
    for k in range(a1.shape[0]):
        i, j = a1[k], a2[k]
        hist1[i] += 1
        hist2[j] += 1
        o +=  (i - j) * (i - j)
    e = 0
    for i in range(max_rat + 1):
        for j in range(max_rat + 1):
            e += hist1[i] * hist2[j] * (i - j) * (i - j)
    e = e / a1.shape[0]
    return 1 - o / e
```

```python
def mse(output, targ): 
    return (output.float() - targ.float()).pow(2).mean()
```

```python
def accuracy(out, yb): 
    return (torch.argmax(out, dim=1)==yb).float().mean()
```

## Training Loop

```python
def fit(epochs: int, model: nn.Sequential, loss_func, opt, train_dl, valid_dl):
    for epoch in range(epochs):
        model.train()
        tot_train_loss = 0.
        for xb, yb in train_dl:
            loss = loss_func(model(xb), yb)
            tot_train_loss += loss
            loss.backward()
            opt.step()
            opt.zero_grad()
            
        model.eval()
        with torch.no_grad():
            tot_loss, tot_mse, tot_acc, tot_acc_qwk = 0., 0., 0., 0.
            for xb, yb in valid_dl:
                pred = model(xb)
                tot_loss += loss_func(pred, yb)
                tot_acc += accuracy(pred, yb)
                tot_acc_qwk += qwk3(pred.argmax(-1), yb)
            nv = len(valid_dl)
            print(f'{epoch} trainloss: {tot_train_loss / len(train_dl)} valid loss: {tot_loss / nv} qwk: {tot_acc_qwk / nv} acc: {tot_acc / nv}')
```

```python
fit(1000, model, loss_func, optimizer, train_dl, valid_dl)
```
